```python
#pretrained models can be called using this pipeline like sentiment-analysis
from transformers import pipeline
classifier = pipeline('sentiment-analysis')
```

```python
classifier('pizza is not that great but the crust is bad')
```

```python
results = classifier(['we are very happy to show you the transformer library.',
                      'we hope you dont hate it.',
                      'still there are things pending'])
for result in results:
  print(f"label: {result['label']}, with score {round(result['score'], 4)}")
```

```python
from transformers import AutoTokenizer, TFAutoModelForSequenceClassification
#whenever we are calling the model we have to call tokenizer bcoz it's different for each model transformer
```

```python
model_name = 'nlptown/bert-base-multilingual-uncased-sentiment'
#this model only exists in Pytorch, so we use the 'from_pt' flah to improt that in tensorflow
model = TFAutoModelForSequenceClassification.from_pretrained(model_name, from_pt= True)
tokenizer= AutoTokenizer.from_pretrained(model_name)
classifier= pipeline('sentiment-analysis', model=model, tokenizer= tokenizer)
```

```python
classifier('I am going to become a doctor')
```

```python
#to check the impact of tokenizer
input = tokenizer('I am going to become a doctor')
print(input)
```

```python
tf_batch= tokenizer([
    'We are very happy to show you the Transformer Library.'
  ,'we hope you dont hate it'
],
                    padding=True,
                    truncation=True,
                    max_length=512,
                    return_tensors='tf')
```
